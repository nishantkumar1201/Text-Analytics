```python
name = 'Fred'

# using the old  .format() method:
print('His name is {var}.'.format(var=name))

# using f-string
print(f'His name is {name}.')
```

```python
print(f'his name is {name!r}')
```

```python
d = {'a':123,'b':456}
print(f'My address: {d['a']} main street')
```

```python
library = [{'author':'Twain','Topic':'rafting','pages':601},{'author':'Feyman','Topic':'physics','pages':95},{'author':'Hamilton','Topic':'mythology','pages':144}]
library
```

```python
library = [('Author','Topic','pages'),('Twain','rafting',601),('Feyman','physics',95),('Hamilton','mythology',144)]
library
```

```python
for book in library:
    print(f'{book[0]:{10}} {book[1]:{8}} {book[2]:{7}}')
```

```python
for book in library:
    print(f'{book[0]:{10}} {book[1]:{8}} {book[2]:*>{7}}')
```

Aliases

```python
with open('test.txt','r') as txt:
    first_line = txt.readlines()[0]

print(first_line)
```

```python
with open('test.txt','r') as txt:
    for line in txt:
        print(line,end='')      #end removes extra line breaks
```

```python
with open('test.txt','r') as txt:
    for line in txt:
        print(line)
```
